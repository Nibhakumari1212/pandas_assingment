# Q5. Average Marks by City

Find the average Marks of each City using **groupby()**.

# Q7. Create Result Column

## Problem Statement
Create a new column named **Result**. If Marks are **40 or more**, store **Pass**, otherwise **Fail**.

## Sample Input
```python
Use the DataFrame from Q1
```

## Sample Output
```python
Updated DataFrame with Result column.
```

# Q8. Remove Duplicate Rows

## Problem Statement
Create the following DataFrame:

| Name | Age | City | Marks |
|------|----:|------|------:|
| Aman | 21 | Bhopal | 85 |
| Rahul | 22 | Indore | 72 |
| Aman | 21 | Bhopal | 85 |
| Neha | 20 | Delhi | 91 |

Remove duplicate rows using **drop_duplicates()**.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Duplicate rows removed.
```

````python
import pandas as pd
d ={
    'Name': ['Aman', 'Rahul', 'Aman', 'Neha'],
    'Age': [21, 22, 21, 20],
    'City': ['Bhopal', 'Indore', 'Bhopal', 'Delhi'],
    'Marks': [85, 72, 85, 91]
}
df = pd.DataFrame(d)
df.drop_duplicates(inplace= True, keep= 'first')
print(df)
````

# Q9. Fill Missing Values

## Problem Statement
Create the following DataFrame:

| Name | Marks |
|------|------:|
| Aman | 85 |
| Rahul | NaN |
| Neha | 91 |
| Priya | NaN |

Fill missing values in the **Marks** column with **0** using **fillna()**.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Missing values replaced with 0.
```

````python
import pandas as pd
import numpy as np
d={
    'Name': ['Aman', 'Rahul', 'Neha', 'Priya'],
    'Marks': [85, np.nan, 91, np.nan]
}
df = pd.DataFrame(d)

df['Marks'] = df['Marks'].fillna(0)
print(df)
````

# Q10. Final Practice

## Problem Statement
Create a DataFrame with the following data:

| Name | City | Marks |
|------|------|------:|
| Aman | Bhopal | 85 |
| Rahul | Indore | 72 |
| Neha | Delhi | 91 |
| Priya | Bhopal | 68 |
| Ravi | Indore | 95 |

Perform all of the following:
1. Display students with Marks > 80.
2. Sort the DataFrame by Marks.
3. Find the average Marks of each City.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Required outputs displayed.
```

````python
import pandas as pd
d = {
    "Name":["Aman","Rahul", "Neha", "Priya", "Ravi"],
    "Age":[21,22,20,23,21],
    "City":["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks":[85,72,91,68,95]
}
df = pd.DataFrame(d)
#1 Display students with marks > 80
print(df[df['Marks'] > 80])
 
#2 sort the DataFrame by marks
print(df.sort_values(by='Marks'))

#3 find the avg marks of each city
print(df.groupby('City')['Marks'].mean())
````
