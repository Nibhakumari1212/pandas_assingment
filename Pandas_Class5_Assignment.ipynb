## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
import pandas as pd
d = {
    "Age":[25,18,30,22]
}
df= pd.DataFrame(d)
print(df.sort_values("Age"))
```

## Q2. Sort Marks in Descending Order

Input:
Marks = [70, 95, 80, 60]

Output:
95
80
70
60

```python
import pandas as pd
d = {
    "Marks":[70,95,80,60]
}
df= pd.DataFrame(d)
print(df.sort_values(by = "Marks",ascending = False))
```

## Q11. Find Correlation

Input:

1.0

```python
import pandas as pd
d = {
    "Maths":[10,20,30,40],
    "Science":[15,25,35,45]
}
df= pd.DataFrame(d)
print(df.corr(numeric_only = True))
```

## Q12. Find Covariance

```python
import pandas as pd
d = {
    "Maths":[10,20,30,40],
    "Science":[15,25,35,45]
}
df= pd.DataFrame(d)
print(df.cov(numeric_only = True))
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
import pandas as pd
d = {
    "Name":["Alice", "Bob", "Emma", "Charlie"], 
    "Age":[29,26,22,21],
    "Marks":[60,70,80,90]
}
df = pd.DataFrame(d)
print("Mean Marks: ",df["Marks"].mean())
print("Median Marks", df["Marks"].median())
print("Max Marks", df["Marks"].max())
print("Min Marks",df["Marks"].min())
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
import pandas as pd
d = {
    "Name":['Alice', 'Bob', 'Emma', 'Alex'],
    "Age":[21,19,16,23],
    "Score":[98,65,89,70],
    "Height":[5.3,4.9,5.8,5.5]
        }

df = pd.DataFrame(d)
print(df.sort_values(by="Score", ascending = False))


#find mean age
print("Mean Age",df["Age"].mean())

# find median height
print("Median Height",df["Height"].median())


#Show score Frequency
print(df['Score'].value_counts())
```
