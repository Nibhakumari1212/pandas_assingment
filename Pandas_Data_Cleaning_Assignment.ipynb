## Q1. Find Missing Values

Input:
df = pd.DataFrame({
    "Name":["Aman","Riya",None],
    "Age":[21,np.nan,25]
})

Output:
    Name    Age
0  False  False
1  False   True
2   True  False

```python
import numpy as np
import pandas as pd

d = {"Name": ["Aman", "Riya", None], "Age": [21, np.nan, 25]}

df= pd.DataFrame(d)
print(df.isna())

```

## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
import numpy as np
import pandas as pd
data = {
    "Name": ["Aman", "Raj", "Aman", None],
    "Age": [20, np.nan, 20, 25],  
    "City": ["Bhopal", "Delhi", "Bhopal", "Indore"],
}
df = pd.DataFrame(data)

df["Age"]=df["Age"].fillna(df["Age"].mean())
df=df.drop_duplicates()

print(df)
```
