Display students belonging to Bhopal.




````python
import pandas as pd
d = {
    "Name":["Amit", "Aman", "Ankit", "Sumit"],
    "Age":[22,21,28,29],
    "City":["Delhi", "Bhopal", "Bihar", "Bhopal"],
    "Marks":[89,98,91, 78]
}
df = pd.DataFrame(d)
print(df[df.City == 'Bhopal'])

````

## Problem Statement
Count occurrences of each city.

## Sample Input
```python
City column
```

```python
Frequency count
```


# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
import pandas as pd

d = {
    "Name": ["Amit", "Aman", "Ankit", "Sumit"],
    "Age": [22, 21, 28, 29],
    "City": ["BH", "Mp", "BH", "Hr"],
    "Marks": [89, 98, 91, 78]
}

df = pd.DataFrame(d)

print(df.groupby('City').Marks.mean())
````


# Q10. Multiple Conditions

## Problem Statement
Display students with Marks > 70 and Age > 20.

## Sample Input
```python
Marks and Age
```

## Sample Output
```python
Filtered rows
```


````python
import pandas as pd
d = {
    "Name":["Amit", "Aman", "Ankit", "Sumit"],
    "Age":[22,21,28,29],
    "Marks":[63,98,15, 78]
}
df = pd.DataFrame(d)
print(df[(df["Marks"] > 70) & (df["Age"] > 20)])
````
