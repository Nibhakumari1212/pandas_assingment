# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
import pandas as pd
d = {
    "City":["Bhopal", "Indore", "Bhopal", "Delhi","pj"]
}
df = pd.DataFrame(d)
print(df["City"].unique())
````


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
import pandas as pd
d = {
    "City":["Bhopal", "Bhopal", "Delhi"], 
    "Marks":[80,90,70]
}
df = pd.DataFrame(d)
df.groupby('City')['Marks'].mean()
````


# Q5. Sort Values

## Problem Statement
Sort students by Marks in ascending order.

## Sample Output
```python
Sorted DataFrame
```


````python
import pandas as pd
d = {
    "Name":["Bob", "Emma", "Charlie", "Alice"],
    "Marks":[87,99,46,78]
}
df = pd.DataFrame(d)
print(df.sort_values(by="Marks",ascending =True))
````


# Q6. Sort Values Descending

## Problem Statement
Sort students by Marks in descending order.

## Sample Output
```python
Sorted DataFrame
```


````python
import pandas as pd
d = {
    "Name":["Bob", "Emma", "Charlie", "Alice"],
    "Marks":[87,99,46,78]
}
df = pd.DataFrame(d)
print(df.sort_values(by="Marks", ascending= False))
````


# Q7. Filter Records

## Problem Statement
Display students having marks greater than 75.

## Sample Output
```python
Filtered rows
```


````python
import pandas as pd
d = {
    "Name":["Bob", "Emma", "Charlie", "Alice"],
    "Marks":[87,99,46,78]
}
df = pd.DataFrame(d)
print(df[df["Marks"]>75])
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
import pandas as pd
d = {
    "Name":["Bob", "Emma", "Charlie", "Alice"],
    "Age":[20,32,22,23],
    "Marks":[87,99,46,78]
}
df = pd.DataFrame(d)
print(df[(df["Marks"] > 75) & (df["Age"] > 20)])
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
import pandas as pd

d ={
    "Name": ["Aman", "Raj","Bob","David"],
    "Department": ["IT", "HR","IT", "HR"],
    "Salary": [40000,  40000,80000,10000]
}
df = pd.DataFrame(d)

print(df.groupby("Department")["Salary"].sum())
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
import pandas as  pd 
d = {
    "Name":["Bob", "Emma", "Charlie", "Alice"],
    "Age": [20,32,22,23],
    "City":["Bhopal", "Bhopal", "Pune", "Indore"],
    "Marks":[81,99,46,78]
}
df = pd.DataFrame(d)
print(df)

# show unique cities
print(df["City"].unique())

# count city frequency
print(df["City"].value_counts())

# filter marsk > 80
print(df[df["Marks"] > 80])

# sort marks
print(df.sort_values(by = "Marks"))
````
